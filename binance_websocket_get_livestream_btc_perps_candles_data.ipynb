```python
import json
import websocket
import pandas as pd
from IPython.display import display, HTML
```

```python
interval = '1m'
pair = 'btcusdt'
contractType = 'perpetual'
socket = f"wss://fstream.binance.com/ws/{pair}_{contractType}@continuousKline_{interval}"
```

```python
t, T, f, L, o, h, l, c, v = [], [], [], [], [], [], [], [], []

def on_message(ws, message):
    msg = json.loads(message)
    bar = msg['k']
    if bar['x'] == False:
      t.append(bar['t'])
      T.append(bar['T'])
      f.append(bar['f'])
      L.append(bar['L'])
      o.append(bar['o'])
      h.append(bar['h'])
      l.append(bar['l'])
      c.append(bar['c'])
      v.append(bar['v'])

def on_error(ws, error):
    print(error)

def on_close(ws, close_status_code, close_msg):
    print("### closed ###")
    bars = {'start_time': t, 'close_time': T, 'first_updated_time': f, 'last_updated_time': L, 'open': o, 'high': h, 'low': l, 'close': c, 'volume': v}
    df = pd.DataFrame.from_dict(bars)
    df['start_time'] = pd.to_datetime(df['start_time'], unit='ms')
    df['close_time'] = pd.to_datetime(df['close_time'], unit='ms')
    df['first_updated_time'] = pd.to_datetime(df['first_updated_time'], unit='ms')
    df['last_updated_time'] = pd.to_datetime(df['last_updated_time'], unit='ms')
    df.set_index('start_time', inplace=True)
#    df.drop(columns='start_time', inplace=True)
    # print(df)
    display(df)

def on_open(ws):
    print("Opened connection!")

ws = websocket.WebSocketApp(socket,
                            on_open=on_open,
                            on_message=on_message,
                            on_error=on_error,
                            on_close=on_close)
```

```python
#the code will run continuously if not paused or until reaching Binance's specified limit, and the dataframe will appear after pausing.
ws.run_forever()
```
